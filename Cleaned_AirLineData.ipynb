# Loading the Data set

```python
def load_data(file):
    try : 
        data = pd.read_csv(file)
        print(f"{file} Data Loaded Successfully!")
        print("\n Data head : ")
        display(data.head())
        print("\n Basic Information :")
        data.info() 
        print("\n Summary Statistics :")
        display(data.describe())
        print("\n Skewness :")
        print(data.skew(numeric_only = True))
        print("\n Kurtosis :")
        print(data.kurt(numeric_only = True))
        return data
        
    except Exception as e :
        print(f"Error loading the dataset : {e}")
    return None
AirlineData = load_data("Indian Airlines.csv")
```

# Data Cleanig And Preprocessing

```python
def clean_data(data):
    # Duplicate Value Treatment.
    print(f"Total no of rows : {len(data)}")
    print(f"Duplicate Values : {data.duplicated().sum()}")
    data.drop_duplicates( inplace = True)
    print(f"Duplicates removed. Rows Remaining : {len(data)} \n")
    
# Missing value treatment 
    print(f"Missing Values :")
    print(round(data.isnull().sum()/len(data)*100,2),"\n")
    data.dropna(inplace=True)

# Outlier detection and treatment 
    numeric_data = data.select_dtypes(include = np.number)
    for col in numeric_data.columns:
        Q1 = data[col].quantile(0.25)
        Q2 = data[col].quantile(0.5)
        Q3 = data[col].quantile(0.75)
        IQR = Q3 - Q1 
        LW = Q1 - (1.5*IQR)
        UW = Q3 + (1.5*IQR)
        outliers = ((data[col] < LW) | data[col] > UW)
        if outliers.sum() > 0 :
            print(f"{col} has {outliers.sum() / len(data) * 100:.2f}% Outliers")
        else : 
            print(f"{col} Column : Does not have any Outliers")
    return data
```

# Extract Day, Month From Traveldate And Convert Departure, Arraival Time In Hours and Minutes 

```python
def Extract_Day_Month_And_Convert_depTime_Arr_Time(data):
    # Extract Day And Month of Travel
    data["TravelDay"] = pd.to_datetime(data["date"],dayfirst=True).dt.day
    data["TravelMonth"] = pd.to_datetime(data["date"],dayfirst=True).dt.month

    # Convert Derature time in hours and Minutes
    data["dep_time_hrs"] = pd.to_datetime(data["dep_time"],format = "%H:%M").dt.hour
    data["dep_time_min"] = pd.to_datetime(data["dep_time"],format = "%H:%M").dt.minute

    # Convert Derature time in hours and Minutes
    data["Arr_time_hrs"] = pd.to_datetime(data["arr_time"],format = "%H:%M").dt.hour
    data["Arr_time_min"] = pd.to_datetime(data["arr_time"],format = "%H:%M").dt.minute
    return data
    
# Convert TravelTime in Total Minutes
def travel_time_minutes(time):
    time = time.strip()
    hours = 0
    minutes = 0
    if "." in time :
        hours = float(time.split('h')[0])
    else :
        if 'h' in time:
            hours = int(time.split('h')[0])
            if 'm' in time:
                minutes = int(time.split('h')[1].split('m')[0].strip())
        else:
            minutes = int(time.split('m')[0])
    return round((hours * 60) + minutes)
```

```python
def convert_stops(value):
    if "non-stop" in value or "zero" in value:
        return 0 
    elif "1-stop" in value or "one" in value :
        return 1 
    elif "2+-stop" in value or "two_or_more" in value:
        return 2
    else : 
        return None

BusinessData["total_stop"] = BusinessData["stop"].apply(convert_stops)
BusinessData["total_stop"].unique()
```
